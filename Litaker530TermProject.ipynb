## Probability Mass Function(PMF) comparing explanatory variable (Bathrooms) based on two scenarios (High Price and Low Price)

```python
high = df[df.price >= 6000000]
low = df[df.price <= 4750000]
high_pmf = thinkstats2.Pmf(high.bathrooms, label = "High Price")
low_pmf = thinkstats2.Pmf(low.bathrooms, label = "Low Price")
```

```python
width = 0.40
thinkplot.PrePlot(2, cols = 2)
thinkplot.Hist(high_pmf, align = "right", width = width)
thinkplot.Hist(low_pmf, align = "left", width = width)
thinkplot.Config(xlabel = "Amount of Bathrooms", ylabel = "Price", legend = True)

thinkplot.PrePlot(2)
thinkplot.SubPlot(2)
thinkplot.Pmfs([high_pmf, low_pmf])
thinkplot.Config(xlabel = "Amount of Bathrooms", ylabel = "Price", legend = True)
```

## Analytical Distribution(Normal) on an explanatory variable (Bedrooms)

```python
mu, var = thinkstats2.TrimmedMeanVar(df.bedrooms, p=0.01)
print("Mean, Var", mu, var)
sigma = np.sqrt(var)
print("Sigma", sigma)
xs, ps = thinkstats2.RenderNormalCdf(mu, sigma, low = 1, high = 6)

thinkplot.Plot(xs, ps, label="model", color="0.6")

cdf = thinkstats2.Cdf(df.bedrooms, label="data")

thinkplot.PrePlot(1)
thinkplot.Cdf(cdf)
thinkplot.Config(title="Bedrooms Normal Distribution", xlabel="Bedrooms", ylabel="CDF")
```

## Definitions for Covariance, Correlation, and Spearman Correlation to extract values from variables used for scatter plots 

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)
    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)
    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
def SpearmanCorr(xs, ys):
    xs = pd.Series(xs)
    ys = pd.Series(ys)
    return xs.corr(ys, method='spearman')
```

## Using permutation test to test the hypothesis 

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
cleaned = df.dropna(subset=['area', 'price'])
data = cleaned.area.values, cleaned.price.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
print ("The pvalue of this test is", pvalue)
print ("The actual correlation value is", ht.actual)
print ("The largest value seen under the simulation is", ht.MaxTestStat())
```
